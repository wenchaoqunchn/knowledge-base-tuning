````python
from bs4 import BeautifulSoup
import re
from pathlib import Path

def get_first_table(soup):
    # 1. 只保留第一个table标签
    first_table = soup.find('table')    
    # 2. 创建只包含这个table的新文档
    new_soup = BeautifulSoup(features='xml')
    new_soup.append(first_table)
    remove_style_attributes(new_soup)
    return new_soup

def only_get_table(soup):
    # 1. 只保留第一个table标签
    tables = soup.find_all('table')    
    # 2. 创建只包含这个table的新文档
    new_soup = BeautifulSoup(features='xml')
    for tab in tables:
        new_soup.append(tab)
    remove_style_attributes(new_soup)
    return new_soup

def remove_trailing_empty_cells(table):
    """移除表格每行末尾共有的空白单元格"""
    trailing_empty_counts = []
    
    for tr in table.find_all('tr'):
        cells = tr.find_all(['td', 'th'])
        empty_cnt = 0
        
        # 从后向前计算空白单元格数量
        for cell in reversed(cells):
            if not cell.text.strip():
                empty_cnt += 1
            else:
                break
                
        trailing_empty_counts.append(empty_cnt)
    
    if not trailing_empty_counts:
        return 0
    
    common_empty_cnt = min(trailing_empty_counts)
    
    if common_empty_cnt > 0:
        for tr in table.find_all('tr'):
            cells = tr.find_all(['td', 'th'])
            for _ in range(common_empty_cnt):
                if cells:
                    cells[-1].decompose()
                    cells = cells[:-1]
    
    return common_empty_cnt

def flatten_paragraphs(soup, label):
    """去除p标签，只保留文本内容在td中"""
    for l in soup.find_all(label):
        # 获取p标签内的所有文本（包括子孙元素的文本）
        text_content = l.get_text(' ', strip=True)
        
        # 用纯文本替换整个p标签
        l.replace_with(text_content)

def remove_empty_rows(soup):
    """移除表格中的空行并简化空白单元格"""
    for tr in soup.find_all('tr'):
        # 先处理空白单元格
        for cell in tr.find_all(['td', 'th']):
            if not cell.text.strip():
                # 创建新的自闭合标签
                new_cell = soup.new_tag(cell.name)
                # 保留原有属性
                new_cell.attrs = cell.attrs.copy()
                cell.replace_with(new_cell)
        
        # 然后检查是否整行为空
        if not tr.text.strip() or all(cell.name in ['td', 'th'] and not cell.contents 
           for cell in tr.find_all(['td', 'th'])):
            tr.decompose()

def add_basic_table_styles(soup):
    """为表格添加基础样式"""
    for table in soup.find_all('table'):
        table['border'] = '1'
        table['cellspacing'] = '0'
        table['cellpadding'] = '4'

def remove_unnecessary_elements(soup):
    """移除不必要的元素"""
    elements_to_remove = ['style', 'script', 'comment', 'head', 'meta', 'link', 'colgroup', 'col', 'title']
    for element in soup.find_all(elements_to_remove):
        element.decompose()

def remove_style_attributes(soup):
    """移除样式和类属性"""
    for tag in soup.find_all(True):
        attrs = {}
        for attr in ['colspan', 'rowspan']:
            if attr in tag.attrs:
                attrs[attr] = tag[attr]
        tag.attrs = attrs

def prettify(soup, compact = False):
    out = soup.prettify()
    tab = " "

    out = re.sub(
        r"(</?t[rdable].*?>)\n\s+",
        r"\1\n",
        out
    )
    out = re.sub(
        r"\s+(</t[rdable]>)",
        r"\n\1",
        out
    )

    out = re.sub(
        r"(<td.*?>)\n(.*?)\n(</td>)",
        r"\1\2\3",
        out
    )
    out = re.sub(
        r"(<td.*?>)\n(</td>)",
        r"\1\2",
        out
    )
    out = re.sub(
        r"<td",
        tab*2 + r"<td",
        out
    )
    out = re.sub(
        r"(</*tr)",
        tab + r"\1",
        out
    )

    if compact:
        out = re.sub(
            r"(</*)t([drh].*?>)",
            r"\1\2",
            out
        )
    return out
            

def medium_compact(soup):
    output = prettify(soup)
    output = re.sub(r"<\?xml.*?\?>", "", output).strip()
    return output

def high_compact(soup):
    """极简表格压缩方案"""
    output = prettify(soup, compact = True)
    output = re.sub(r"<\?xml.*?\?>", "", output)
    output = r"<!-- 标签映射表={<table>:<t>,<tr>:<r>,<td>:<d>,<th>:<h>} -->" + output
    return output

def simplify_html_table(html_content, compact_level = 0):
    """主函数：简化XHTML表格内容"""
    soup = BeautifulSoup(html_content, 'lxml')

    # 处理
    remove_unnecessary_elements(soup)
    remove_style_attributes(soup)
    flatten_paragraphs(soup,"font")
    flatten_paragraphs(soup,"b")
    remove_empty_rows(soup)
    
    for table in soup.find_all('table'):
        remove_trailing_empty_cells(table)
    
    add_basic_table_styles(soup)

    output = ""
    if compact_level == 0:
        output = re.sub(r'\s+', ' ', str(soup))
    elif compact_level == 1:
        output = medium_compact(only_get_table(soup))
    else:
        output = high_compact(only_get_table(soup))
    
    return output

def html2md(doc_in, out_dir):
    with open(doc_in, 'r', encoding='utf-8') as f:
        html_content = f.read()
    simplified_html = simplify_html_table(html_content, compact_level=1)
    md = "```markdown\n" + simplified_html + "\n```"
    with open(out_dir / doc_in.with_suffix(".md").name, 'w', encoding='utf-8') as f:
        f.write(md)
    return 

out_dir = Path("./data/out/table")
cnt = 0
for html  in Path("./data/mid").glob("*.html"):
    cnt += 1
    print(cnt)
    html2md(html, out_dir)
    
````

````python
import pandas as pd
import pdfplumber
import re
import shutil
import camelot

def md_formater(str_in):
    if "章" in str_in:
        str_in = re.sub(
            r"(第[一二三四五六七八九十百]+章)",  # 只匹配数字/中文数字
            r"\n\n## \1 ",
            str_in,
        )
    if "条" in str_in:
        str_in = re.sub(
            r'(第[一二三四五六七八九十百]+条)',
            r"\n\n#### \1\n",
            str_in,
        )
    if "节" in str_in:
        str_in = re.sub(
            r'(第[一二三四五六七八九十百]+节)',
            r"\n\n### \1 ",
            str_in,
        )
    if "（" in str_in:
        str_in = re.sub(
            r'(（[一二三四五六七八九十百]+）)',
            r"\n\n\1",
            str_in,
        )
    if "." in str_in:
        str_in = re.sub(
            r'(\d+)\.',
            r"\n\n（\1）",
            str_in,
        )
    if "附件" in str_in:
        str_in = re.sub(
            r"(附件：)",
            r"\n\n## \1\n",
            str_in,
        )
    return str_in

def format_table(table):
    df = pd.DataFrame(table)
    df = df.fillna("")
    df = df.map(lambda str_in: "".join(str_in.split("\n")))
    md = df.to_markdown(index=False)
    md = re.sub(
        r"( {2,})",
        r" ",
        md
    )
    md = re.sub(
        r"(-{3,})",
        r"-----",
        md
    )
    intent = " "*0
    md = intent + f"\n{intent}".join(md.split("\n"))
    return f"\n{intent}\n" + md + f"\n{intent}\n"

def find_non_empty_indexes(inserted_list, table):
    arr = [item for row in table for item in row]
    first = None
    # print(table)
    try:
        for i in range(len(inserted_list)):
            if inserted_list[i] is not None:
                # print(inserted_list[i])
                if inserted_list[i].strip() != "" and inserted_list[i] == table[0][0]:
                    first = i
                    break
                
            # 寻找第一个非空字符串
        for i in range(len(inserted_list)):
            if inserted_list[i] is not None:
                # print(inserted_list[i])
                if inserted_list[i].strip() != "" and inserted_list[i] == table[0][0] and inserted_list[i+1] == table[0][1]:
                    first = i
                    break
    except:
        return first
    return first

def replace_table_in_text(tables, text_list):
    inserted_list = text_list

    for table in tables:
        # print(table)
        table_inserted = format_table(table)
        # print(table_inserted)
        h_idx = find_non_empty_indexes(inserted_list, table)
        if h_idx is not None:
            inserted_list = text_list[:h_idx] + [table_inserted] + text_list[h_idx:]
        else:
            inserted_list = text_list
        # inserted_list = inserted_list + f"\n\n{table_inserted}"

    return inserted_list

def pdfdoc2md(doc_in, out_path):
    pdf_path = doc_in

    with pdfplumber.open(pdf_path) as pdf:
        page_list = []
        for page in pdf.pages:
            num = page.page_number
            word = page.extract_words(x_tolerance=3, y_tolerance=3)
            tables = page.extract_tables()
            text_list = [w.get("text") for w in word]
            # if num == 3:
            #     print(text_list)
            text_list = replace_table_in_text(tables, text_list)
            if len(text_list):
                text_list.pop()
            text_list = [md_formater(l) for l in text_list]

            # print(text_list)
            page_list.append("".join(text_list))
        # print("".join(page_list))
        content = "# " + "".join(page_list)
    with open(out_path, "w", encoding="utf-8") as f:
        f.write(content)


def pdftab2md(pdf_path, out_path):
    ctabs= camelot.io.read_pdf(pdf_path, pages='1',flavor='lattice',strip_text = '\n')####从1开始计数
    md = f"# {pdf_path.stem}\n\n" + "\n\n".join([ctab.df.to_markdown(index=False) for ctab in ctabs])
    
    md = re.sub(
        r"( {2,})",
        r" ",
        md
    )
    md = re.sub(
        r"(-{3,})",
        r"-----",
        md
    )
    with open(out_path, "w", encoding="utf-8") as f:
        f.write(md)




````

````python
from pathlib import Path
from typing import List

def merge_md_files(pdf_tab_dir: Path, llm_tab_dir: Path, merge_tab_dir: Path):
    """
    拼接pdf_tab和llm_tab中的同名MD文件，结果存入merge_tab
    
    参数:
        pdf_tab_dir: 包含PDF生成MD文件的目录
        llm_tab_dir: 包含LLM生成MD文件的目录
        merge_tab_dir: 存放合并结果的目录
    """
    # 确保输出目录存在
    merge_tab_dir.mkdir(parents=True, exist_ok=True)
    
    # 获取pdf_tab中的所有md文件
    pdf_files = list(pdf_tab_dir.glob("*.md"))
    
    # 遍历所有pdf文件
    for pdf_file in pdf_files:
        # 获取对应的llm文件路径
        llm_file = llm_tab_dir / pdf_file.name
        
        # 检查llm文件是否存在
        if not llm_file.exists():
            print(f"警告: {pdf_file.name} 在llm_tab中不存在，跳过")
            continue
            
        # 读取两个文件内容
        pdf_content = pdf_file.read_text(encoding="utf-8")
        llm_content = llm_file.read_text(encoding="utf-8")
        
        # 拼接内容(pdf在前，llm在后)
        merged_content = f"{pdf_content}\n\n---\n\n{llm_content}"
        
        # 写入合并后的文件
        output_file = merge_tab_dir / pdf_file.name
        output_file.write_text(merged_content, encoding="utf-8")
        print(f"已合并: {pdf_file.name}")

def main():
    # 定义目录路径
    base_dir = Path("./data/out")  # 当前目录，可根据需要修改
    pdf_tab = base_dir / "pdf_tab"
    llm_tab = base_dir / "llm_tab"
    merge_tab = base_dir / "merge_tab"
    
    # 验证输入目录存在
    if not pdf_tab.exists() or not pdf_tab.is_dir():
        raise FileNotFoundError(f"目录不存在: {pdf_tab}")
    if not llm_tab.exists() or not llm_tab.is_dir():
        raise FileNotFoundError(f"目录不存在: {llm_tab}")
    
    # 执行合并
    merge_md_files(pdf_tab, llm_tab, merge_tab)
    print("所有文件合并完成！")

if __name__ == "__main__":
    main()
````

````python
import yaml
from pathlib import Path
import json
class IdKeyPair:
    def mask(value):
        return value[:2] + '****' + value[-2:] if value and len(value) > 4 else '****'
            
    def __init__(self, dic):
        self.id = dic.get("id")
        self.key = dic.get("key")
    def __str__(self):
        return (
            f"  id: {mask(self.id)}\n"
            f"  key: {mask(self.key)}\n"
        )

class Prompts:
    def __init__(self, dic):
        self.start_prompt = dic.get("start")
        self.continue_prompt = dic.get("continue")
    def __str__(self):
        return (
            f"  start: {self.start_prompt}\n"
            f"  continue: {self.continue_prompt}\n"
        )

class Config:
    def __init__(self, config_path):
        # 支持YAML和JSON两种格式
        with open(config_path, "r") as file:
            if config_path.suffix in ['.yml', '.yaml']:
                self.__config_dict__ = yaml.safe_load(file)
            elif config_path.suffix == '.json':
                self.__config_dict__ = json.load(file)
        
        self.url = self.__config_dict__.get("url")
        self.dataset = IdKeyPair(self.__config_dict__.get("dataset"))
        self.app = IdKeyPair(self.__config_dict__.get("app"))
        self.prompts = Prompts(self.__config_dict__.get("prompts"))
        
    def __str__(self):
        return (
            f"dataset: \n{self.dataset}"
            f"app: \n{self.app}"
            f"prompts: \n{self.prompts}"
        )
````

````python
import re

def read_md(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        md_str = file.read()
    return md_str

def remove_spire(text):
    # 去除特定字符串内容
    cleaned_text = text.replace("Evaluation Warning: The document was created with Spire.Doc for Python.", "")
    
    # # 移除多余连续换行符
    # lines = cleaned_text.split('\n')
    # cleaned_lines = [line for line in lines if line.strip()]
    
    return cleaned_text

def remove_style(text):
    cleaned_text = text.replace("**", "").replace("#", "")
    cleaned_text = remove_multi_newlines(cleaned_text)
    return cleaned_text

def change_bullet_points(text):
    # 定义正则表达式模式
    pattern = r'(\d+)\.\s'
    replacement = r'\1） '
    
    # 使用 re.sub 进行替换
    modified_text = re.sub(pattern, replacement, text)
    
    return modified_text    


def mask(key):
    return "*****" + key[-4:]

def remove_think(input_string):
    # 匹配形如 <think>\n\n 其他任意字符</think>\n\n 的子串
    pattern = re.compile(r'<think>.*?</think>', re.DOTALL)
    
    # 删除匹配到的子串
    cleaned_string = re.sub(pattern, '', input_string)
    
    return cleaned_string

def remove_multi_newlines(input_string):
    # 匹配两个或更多连续的换行符
    pattern = re.compile(r'\n{3,}')
    
    # 替换匹配到的部分为单个换行符
    cleaned_string = re.sub(pattern, '\n\n', input_string)
    
    stripped_string = cleaned_string.strip()
    
    return stripped_string



def clean_content(content):
    return remove_multi_newlines(remove_think(content))
````

````python
import json
from pathlib import Path

def load_json_to_dict(file_path):
    """安全地将JSON文件加载为字典"""
    try:
        json_path = Path(file_path)
        if not json_path.exists():
            raise FileNotFoundError(f"文件不存在: {file_path}")
        
        return json.loads(json_path.read_text(encoding='utf-8'))
    except json.JSONDecodeError as e:
        print(f"JSON解析错误: {e}")
        return None
    except Exception as e:
        print(f"发生错误: {e}")
        return None
````
